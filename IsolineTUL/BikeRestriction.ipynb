```python
import json

with open ('apiReturn.json', 'r') as fp:
    data = json.load(fp)

geoJsonCells = [ {
            "type": "Feature",
            "properties": {
                "Cell": 'IsolineTUL'
            },
            "geometry": {
                "type": "Polygon",
                "coordinates": [[ [ float(i.split(',')[1]), float(i.split(',')[0]) ] for i in data['response']['isoline'][0]['component'][0]['shape'] ]]
    }} ]

with open('IsolineTUL'+'.geojson', 'w') as fp:
    json.dump({
        "type": "FeatureCollection",
        "name": "ISOLINETUL",
        "crs": { "type": "name", "properties": { "name": "ISOLINETUL" } },
        "features": geoJsonCells}
    , fp)
```
